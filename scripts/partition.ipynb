```python
import numpy as np
from BM import BoltzmannMachine
import matplotlib.pyplot as plt

# Params
ns = list(range(1, 10))
p = 50 # number of patterns
eta = .005 # learning rate
epsilon = 1e-13 # convergence criterion
num_iter = 1000 # number of MH steps
```

```python
exacts = []
approxs = []
for n in ns:
    x_small = np.random.choice([-1, 1], size=(n,p))
    w_init = np.ones((n,n))
    w = w_init = w_init - np.diag(w_init)
    theta = theta_init = np.random.randn(n)

    # Init Boltzmann Machine
    bm = BoltzmannMachine(x_small, w, theta, eta, epsilon, num_iter, approx_partition=False)
    exacts.append(bm.partition_function())
    bm = BoltzmannMachine(x_small, w, theta, eta, epsilon, num_iter, approx_partition=True)
    approxs.append(bm.partition_function())

plt.plot(ns, np.log(approxs), label='Approx')
plt.plot(ns, np.log(exacts), label='Exact')
plt.legend()
plt.grid()
plt.xlabel('Spins')
plt.ylabel('Log Partition Function')
plt.title(f'Comparison of Exact And MF Approximation \n of Log Partition Function(Not Converged)')
```

- Approximation is lower then the exact partition function, which matches the theoretical consideration.
- The deviation between approximation and error increases as the amount of spins gets worse. 

```python
exacts = []
approxs = []
for n in ns:
    x_small = np.random.choice([-1, 1], size=(n,p))
    w_init = np.ones((n,n))
    w = w_init = w_init - np.diag(w_init)
    theta = theta_init = np.random.randn(n)

    # Init Boltzmann Machine
    bm = BoltzmannMachine(x_small, w, theta, eta, epsilon, num_iter, approx_partition=False)
    bm.train('exact')
    exacts.append(bm.partition_function())
    bm = BoltzmannMachine(x_small, w, theta, eta, epsilon, num_iter, approx_partition=True)
    approxs.append(bm.partition_function())

plt.plot(ns, np.log(approxs), label='Approx')
plt.plot(ns, np.log(exacts), label='Exact')
plt.legend()
plt.grid()
plt.xlabel('Spins')
plt.ylabel('Log Partition Function')
plt.title(f'Comparison of Exact And MF Approximation \n of Log Partition Function(Converged)')
```

- Approximation is lower then the exact partition function, which matches the theoretical consideration.
- When converging the approximations are better.
- The deviation between approximation and error increases as the amount of spins gets worse.
